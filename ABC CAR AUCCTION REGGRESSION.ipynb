```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('ABC auctions.csv')
```

```python
# Define the dependent and independent variables
y = data['earningspct']
X = data[['legacy', 'sold','ftturn', 
          'ptturn', 'ohsapercar', 
          'lotdampercar']]
```

```python
# Check for multicollinearity among the independent variables
corr_matrix = X.corr()
print(corr_matrix)

sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, square=True)

# Add title and axis labels
plt.title('Correlation Matrix')
plt.xlabel('Variables')
plt.ylabel('Variables')

# Show the plot
plt.show()
```

```python
# Run a multiple regression model FIRST REGESSION 
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
#Second Reggresion 
# Step 2: Define the dependent and independent variables
y = data['ftturn']
X = data[['legacy','sold','ei', 'oc']]
```

```python
# Step 4: Run a multiple regression model
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```
